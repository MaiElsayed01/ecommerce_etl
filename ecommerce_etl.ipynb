```python
payment_df = pd.read_csv('recource/payment_dataset.csv')
payment_df.info()
payment_df.head()
```

```python
feedback_df = pd.read_csv('recource/feedback_dataset.csv')
feedback_df.info()
feedback_df.head()
```

```python
order_df = pd.read_csv('recource/order_dataset.csv')
order_df.info()
order_df.head()
```

```python
order_item_df = pd.read_csv('recource/order_item_dataset.csv')
order_item_df.info()
order_item_df.head()
```

Customer Dimension

```python
# change the data type of the customer_zip_code to string
customer_df['customer_zip_code'] = customer_df['customer_zip_code'].astype(str)
customer_df.info()
```

```python
# check for duplicates
customer_df[customer_df.duplicated(subset=['user_name'])]
```

```python
# add a user_key (int) column
customer_df = customer_df.reset_index()
customer_df = customer_df.rename(columns={'index': 'customer_key'})
```

```python
# rename columns
customer_df = customer_df.rename(columns={'user_name': 'customer_id'})
```

```python
customer_df.head()
```

Seller Dimension

```python
seller_df['seller_zip_code'] = seller_df['seller_zip_code'].astype(str)
seller_df.info()
```

```python
# check for duplicates
seller_df[seller_df.duplicated(subset=['seller_id'])]
```

```python
# add a seller_key (int) column
seller_df = seller_df.reset_index()
seller_df = seller_df.rename(columns={'index': 'seller_key'})
```

```python
seller_df.head()
```

Product Dimension

```python
# add a product_key (int) column
product_df = product_df.reset_index()
product_df = product_df.rename(columns={'index': 'product_key'})

# rename columns
product_df = product_df.rename(columns={'product_name_lenght': 'product_name_length', 'product_description_lenght': 'product_description_length'})

product_df.head()
```

Date Dimension

```python
start_date = order_df['order_date'].min()
end_date = order_df['order_date'].max()

date_df = pd.DataFrame(pd.date_range(start_date, end_date), columns=['full_date'])
date_df['year'] = date_df['full_date'].dt.year
date_df['quarter'] = date_df['full_date'].dt.quarter
date_df['month'] = date_df['full_date'].dt.month
date_df['day'] = date_df['full_date'].dt.day
date_df['weekday'] = date_df['full_date'].dt.weekday + 1
date_df['day_name'] = date_df['full_date'].dt.day_name()
date_df['month_name'] = date_df['full_date'].dt.month_name()
date_df['season'] = date_df['month'].apply(
    lambda month:
        'Winter' if 11 <= month<= 2
        else 'Spring' if 3 <= month <= 5
        else 'Summer' if 6 <= month <= 8
        else 'Fall'
)
date_df['full_date'] = date_df['full_date'].dt.date
date_df = date_df.reset_index()
date_df = date_df.rename(columns={'index': 'date_key'})

date_df.head()
```

Time Dimension

```python
times = [datetime.time(h,m,s) for h in range(24) for m in range(60) for s in range(60)]

time_df = pd.DataFrame(times, columns=['full_time'])
time_df['hour'] = time_df['full_time'].apply(lambda time: time.hour)
time_df['minute'] = time_df['full_time'].apply(lambda time: time.minute)
time_df['second'] = time_df['full_time'].apply(lambda time: time.second)
time_df['period_of_day'] = time_df['hour'].apply(
    lambda hour:
        'Morning' if 6 <= hour <= 11
        else 'Afternoon' if 12 <= hour <= 17
        else 'Evening' if 18 <= hour <= 23
        else 'Night'
)

time_df = time_df.reset_index()
time_df = time_df.rename(columns={'index': 'time_key'})

time_df.head()
```

Order Fact

```python
#handle date and time columns
order_df['order_date'] = pd.to_datetime(order_df['order_date'])
order_df['order_time'] = order_df['order_date'].dt.time
order_df['order_date'] = order_df['order_date'].dt.date

order_df['delivered_date'] = pd.to_datetime(order_df['delivered_date'])
order_df['delivered_time'] = order_df['delivered_date'].dt.time
order_df['delivered_date'] = order_df['delivered_date'].dt.date

order_df['estimated_time_delivery'] = pd.to_datetime(order_df['estimated_time_delivery']) 
order_df['estimated_time_delivery'] = order_df['estimated_time_delivery'].dt.date
order_df = order_df.rename(columns={'estimated_time_delivery': 'estimated_delivery_date'})

order_df.info()
```

```python
# add order payments to the order dataframe
order_df = order_df.merge(
    payment_df[['order_id', 'payment_type', 'payment_installments', 'payment_value']],
    on='order_id',
    how='left'
)
order_df.head()
```

```python
# add order feedback to the order dataframe
order_df = order_df.merge(
    feedback_df[['feedback_score', 'order_id', 'feedback_form_sent_date', 'feedback_answer_date']],
    on='order_id',
    how='left'
)
# rename columns of feedback
order_df = order_df.rename(columns={
    'feedback_form_sent_date': 'feedback_date',
    'feedback_answer_date': 'feedback_answer_date'
})
# change the data type of the feedback_date and feedback_answer_date to date
order_df['feedback_date'] = pd.to_datetime(order_df['feedback_date']).dt.date
order_df['feedback_answer_date'] = pd.to_datetime(order_df['feedback_answer_date']).dt.date

order_df.head()
```

```python
# get total price of each order and join with order dimension
order_item_df['total_item_value'] = order_item_df['price'] + order_item_df['shipping_cost']  
total_order_value = order_item_df.groupby('order_id')['total_item_value'].sum().reset_index().rename(columns={'total_item_value': 'total_order_value'})

total_order_value.head()

order_df = order_df.merge(
	total_order_value,
	on='order_id',
	how='left'
)

# get order item count per order and join with order dimension
order_item_count = order_item_df.groupby('order_id')['order_item_id'].count().reset_index().rename(columns={'order_item_id': 'order_item_count'})
order_df = order_df.merge(
	order_item_count,
	on='order_id',
	how='left'
)
order_df.head()
```

```python
# merge the order and customer dataframes
order_df = pd.merge(order_df, customer_df[['customer_id', 'customer_key']], left_on= 'user_name', right_on= 'customer_id', how='inner')
# drop the user_name and user_id columns
order_df = order_df.drop(columns=['user_name'])
order_df = order_df.drop(columns=['customer_id'])
order_df.head()
```

```python
# remove pickup and approved columns from the order dataframe
order_df = order_df.drop(columns=['order_approved_date','pickup_date'])
order_df.head()
```

```python
# add date key for all columns in order_df

#for order_date
order_df = pd.merge(order_df, date_df[['date_key','full_date']], left_on='order_date', right_on='full_date', how='inner')
order_df = order_df.drop(columns=['order_date'])
order_df = order_df.drop(columns=['full_date'])
order_df = order_df.rename(columns={'date_key': 'order_date_key'})

#for delivered_date
order_df = pd.merge(order_df, date_df[['date_key','full_date']], left_on='delivered_date', right_on='full_date', how='inner')
order_df = order_df.drop(columns=['delivered_date'])
order_df = order_df.drop(columns=['full_date'])
order_df = order_df.rename(columns={'date_key': 'delivered_date_key'})

# for estimated_delivery_date
order_df = pd.merge(order_df, date_df[['date_key','full_date']], left_on='estimated_delivery_date', right_on='full_date', how='inner')
order_df = order_df.drop(columns=['estimated_delivery_date'])
order_df = order_df.drop(columns=['full_date'])
order_df = order_df.rename(columns={'date_key': 'estimated_delivery_date_key'})

# for feedback_date
order_df = pd.merge(order_df, date_df[['date_key','full_date']], left_on='feedback_date', right_on='full_date', how='inner')
order_df = order_df.drop(columns=['feedback_date'])
order_df = order_df.drop(columns=['full_date'])
order_df = order_df.rename(columns={'date_key': 'feedback_date_key'})

# for feedback_answer_date
order_df = pd.merge(order_df, date_df[['date_key','full_date']], left_on='feedback_answer_date', right_on='full_date', how='inner')
order_df = order_df.drop(columns=['feedback_answer_date'])
order_df = order_df.drop(columns=['full_date'])
order_df = order_df.rename(columns={'date_key': 'feedback_answer_date_key'})

order_df.head()
```

```python
# add time key for all columns in order_df

# for order_time
order_df = pd.merge(order_df, time_df[['time_key','full_time']], right_on= 'full_time', left_on='order_time', how='left')
order_df = order_df.drop(columns=['order_time'])
order_df = order_df.drop(columns=['full_time'])
order_df = order_df.rename(columns={'time_key': 'order_time_key'})

# for delivered_time
order_df = pd.merge(order_df, time_df[['time_key','full_time']], right_on= 'full_time', left_on='delivered_time', how='left')
order_df = order_df.drop(columns=['delivered_time'])
order_df = order_df.drop(columns=['full_time'])
order_df = order_df.rename(columns={'time_key': 'delivered_time_key'})

order_df.head()
```

```python
order_df = order_df.reset_index()
order_df = order_df.rename(columns={'index': 'order_key'})
order_df.head()
```

Order Item Fact

```python
# add user key to order_item_df
order_item_df = pd.merge(order_item_df, order_df[['order_id', 'customer_key']], on='order_id', how='inner')

# add product key to order_item_df
order_item_df = pd.merge(order_item_df, product_df[['product_id', 'product_key']], on='product_id', how='inner')
order_item_df = order_item_df.drop(columns=['product_id'])

# add seller key to order_item_df
order_item_df = pd.merge(order_item_df, seller_df[['seller_id', 'seller_key']], on='seller_id', how='inner')
order_item_df = order_item_df.drop(columns=['seller_id'])

order_item_df.head()
```

```python
# add date and time columns for pickup
order_item_df['pickup_limit_date'] = pd.to_datetime(order_item_df['pickup_limit_date'])
order_item_df['pickup_limit_time'] = order_item_df['pickup_limit_date'].dt.time
order_item_df['pickup_limit_date'] = order_item_df['pickup_limit_date'].dt.date

# add date key for pickup
order_item_df = pd.merge(order_item_df, date_df[['date_key','full_date']], left_on='pickup_limit_date', right_on='full_date', how='inner')
order_item_df = order_item_df.drop(columns=['pickup_limit_date'])
order_item_df = order_item_df.drop(columns=['full_date'])
order_item_df = order_item_df.rename(columns={'date_key':'pickup_limit_date_key'})

# add time key for pickup
order_item_df = pd.merge(order_item_df, time_df[['time_key','full_time']], left_on='pickup_limit_time', right_on='full_time', how='inner')
order_item_df = order_item_df.drop(columns=['pickup_limit_time'])
order_item_df = order_item_df.drop(columns=['full_time'])
order_item_df = order_item_df.rename(columns={'time_key':'pickup_limit_time_key'})  

order_item_df.head()
```

```python
# drop total_item_value column
order_item_df = order_item_df.drop(columns=['total_item_value'])
order_item_df.head()
```

```python
# add order key to order_item_df
order_item_df = order_item_df.reset_index()
order_item_df = order_item_df.rename(columns={'index': 'order_item_key'})
order_item_df.head()
```

Load Data into DWH

```python
from sqlalchemy import create_engine
from config import server, database, driver 

connection_string = f"mssql+pyodbc://@{server}/{database}?driver={driver}"

engine = create_engine(connection_string)

database
```

```python


tables = {
    'customer_dim': customer_df,
    'seller_dim': seller_df,
    'product_dim': product_df,
    'date_dim': date_df,
    'time_dim': time_df,
    'order_fact': order_df,
    'order_item_fact': order_item_df
}
try:
    for name, df in tables.items():
        df.to_sql(
            name=name,
            con=engine,
            if_exists='append', 
            index=False,
            chunksize=1000     
        )
        print(f"{name} inserted successfully")
    
    print("All tables inserted!")
except Exception as e:
    print(f"Error inserting data: {e}")
```

```python
# close the connection
engine.dispose()
```
